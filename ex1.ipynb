تمرین ۱ – ماسک کردن یک رنگ

یک عکس رنگی انتخاب کن.

یک کانال رنگ (R، G یا B) را انتخاب کن.

تمام پیکسل‌هایی که مقدار آن کانال بیشتر از یک حد مشخص است را با رنگ مشکی جایگزین کن.

نتیجه را همراه با تصویر اصلی نمایش بده.

هدف: تمرین کار با cv2.split(), cv2.merge(), slicing و masking.

</div>

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('pic/color2.png')
b,g,r = cv2.split(img)
r.shape
```

```python
r.shape[0]*r.shape[1]
```

```python
threshold = 100
print(np.count_nonzero(r[r>threshold]))
r[r>threshold] = 0
print(np.count_nonzero(r[r>threshold]))
```

```python
img_changed = cv2.merge([b,g,r])
plt.figure(figsize=(12,5))
plt.subplot(121)
plt.imshow(img[...,::-1])
plt.title('original image')
plt.subplot(122)
plt.imshow(img_changed[...,::-1])
plt.title('changed image')
plt.tight_layout()
plt.show()
```
